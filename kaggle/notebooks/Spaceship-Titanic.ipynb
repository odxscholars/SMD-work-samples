# Figure out which features are important in predicting whether or not a passenger gets transported

### Home Planet

```python
df_arr = []
train_data['HomePlanet'].unique()

for i in train_data["HomePlanet"].unique():
    df_arr.append(tuple((i,train_data.loc[train_data.HomePlanet==i]["Transported"])))
    
    
for i in df_arr:
    print(i[0])
    print(i[1].value_counts())
    print("\n")
    
for i in df_arr:
    try:
        print("for planet: ",i[0], " the transport rate is: ", (sum(i[1])/len(i[1])) * 100)
        print(len(i[1]))
    except:
        #print error if there is a division by zero
        print("error")
        
    
```

from sklearn.ensemble import HistGradientBoostingClassifier

```python
from sklearn.ensemble import HistGradientBoostingClassifier
y = train_data["Transported"]
features = ["HomePlanet", "CryoSleep", "Destination", "VIP"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = HistGradientBoostingClassifier()
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Transported': predictions})
output.to_csv('Datasets/Spaceship-Titanic/my_submission.csv', index=False)
```
